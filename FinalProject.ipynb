# DATA CLEANING

***CHECK NULL***

```python
df_customers.isnull().sum()
```

```python
df_orders.isnull().sum()
```

```python
nan_df_orders = df_orders[df_orders.isna().any(axis=1)]
nan_df_orders
```

```python
df_orders=df_orders.dropna(subset=['order_approved_at','order_delivered_carrier_date','order_delivered_customer_date']).reset_index(drop = True)
df_orders
```

```python
df_orders.isnull().sum()
```

```python
df_order_items.isnull().sum()
```

```python
df_products.isnull().sum()
```

```python
nan_df_products = df_products[df_products.isna().any(axis=1)]
nan_df_products
```

```python
df_products=df_products.dropna(subset=['product_category_name','product_name_lenght','product_description_lenght','product_photos_qty','product_weight_g','product_length_cm','product_height_cm','product_width_cm']).reset_index(drop = True)
df_products
```
